```python
n = int(input()) 
count = 0

while n >= 0:   # 3kg씩 계속 빼다가 0이하가 되면 정확히 만들 수 없는 경우       18 > 15 > ...12 ..> 
    if n % 5 == 0:
        count += n // 5  # 15면 5나눠서 몫  3 증가 
        print(count)
        break
    
    n -= 3
    count += 1
else:
    print(-1)
    
    
```
